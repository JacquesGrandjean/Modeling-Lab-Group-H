# 1. Three basic assumptions of the Langmuir model:
     - The surface with the containing sites on it is perfectly flat with no corrugations. In additon, all sites on the surface are energitically equivalent. 
     - Each site can hold only one molecule.
     - There are no interactions between absrobed molecules.

# 2. Prove.
     - At high pressure, the gas uptake reaches a plateau.
We start from the Langmuir equation:

$$
q(P) = q_{\max}\,\frac{K(T)P}{1+K(T)P}
$$

where  
- $q(P)$ = adsorbed amount at pressure $P$,  
- $q_{\max} > 0$ = maximum adsorption capacity (monolayer coverage),  
- $K(T) > 0$ = affinity constant.  


Step 1: Rewrite fraction
Divide numerator and denominator by $P$:

$$
q(P) = q_{\max}\,\frac{K(T)}{\tfrac{1}{P}+K(T)}.
$$


Step 2: Limit at high pressure
As $P \to \infty$, we have $\tfrac{1}{P} \to 0$, so:

$$
\lim_{P \to \infty} q(P) = q_{\max}\,\frac{K(T)}{0+K(T)} = q_{\max}.
$$


Step 3: Monotonicity
Differentiate with respect to $P$:

$$
\frac{dq}{dP} = q_{\max}\,\frac{K(T)}{(1+K(T)P)^2} > 0 \quad \forall P \ge 0.
$$

Thus $q(P)$ is strictly increasing, but approaches a finite upper bound.

    - At low pressure, the gas uptake is a linear function of pressure

We consider the Langmuir equation:

$$
q(P) = q_{\max}\,\frac{K(T)P}{1+K(T)P}
$$


Step 1: Behavior at low pressure
When $P \to 0$, the denominator can be expanded:

$$
1 + K(T)P \approx 1 \quad \text{(since $K(T)P \ll 1$)}.
$$

Thus:

$$
q(P) \approx q_{\max}\,(K(T)P).
$$


Step 2: Linear dependence
This shows that:

$$
q(P) \sim (q_{\max} K(T))\,P \quad \text{as } P \to 0.
$$


Conclusion
At low pressure, the adsorbed amount $q(P)$ is **directly proportional to pressure**.  
The proportionality constant is the **Henry constant**:

$$
K_H = q_{\max} K(T).
$$

# 3. Deduce the temperature dependance of the Henry's coefficient

Temperature dependence of the Henry coefficient

From the low-pressure limit of the Langmuir isotherm we have the Henry constant:

$$
K_H \;=\; q_{\max}\,K(T),
$$

where  

- $q_{\max}$ = monolayer capacity  
- $K(T)$ = Langmuir affinity (equilibrium) constant  

Thermodynamic temperature dependence of $b$

Since $K(T)$ is an equilibrium constant, it follows the usual thermodynamic relation:

$$
K(T)=\exp\!\left(-\tfrac{\Delta G_{\mathrm{ads}}}{RT}\right)
=\exp\!\left(\tfrac{\Delta S_{\mathrm{ads}}}{R}\right)\,\exp\!\left(-\tfrac{\Delta H_{\mathrm{ads}}}{RT}\right),
$$

where  

- $\Delta G_{\mathrm{ads}}$ = Gibbs free energy of adsorption  
- $\Delta H_{\mathrm{ads}}$ = enthalpy change of adsorption  
- $\Delta S_{\mathrm{ads}}$ = entropy change of adsorption  
- $R$ = gas constant  
- $T$ = absolute temperature  

Expression for $K_H(T)$

Therefore:

$$
K_H(T)=q_{\max}(T)\,\exp\!\left(\tfrac{\Delta S_{\mathrm{ads}}}{R}\right)\,
\exp\!\left(-\tfrac{\Delta H_{\mathrm{ads}}}{RT}\right).
$$

Taking logarithms gives a van ’t Hoff–like form:

$$
\ln K_H(T)=\ln q_{\max}(T)+\frac{\Delta S_{\mathrm{ads}}}{R}
-\frac{\Delta H_{\mathrm{ads}}}{RT}.
$$

Common simplification

If $q_{\max}$ is approximately temperature-independent:

$$
K_H(T)=K_0\,\exp\!\left(-\tfrac{\Delta H_{\mathrm{ads}}}{RT}\right),
\qquad
\ln K_H(T)=\ln K_0-\frac{\Delta H_{\mathrm{ads}}}{RT},
$$

with  

$$
K_0 = q_{\max}^0 \exp\!\left(\tfrac{\Delta S_{\mathrm{ads}}}{R}\right).
$$

Physical interpretation

- If adsorption is **exothermic** ($\Delta H_{\mathrm{ads}}<0$):  
  $K_H$ **decreases** as $T$ increases → adsorption favored at low $T$.  

- If adsorption is **endothermic** ($\Delta H_{\mathrm{ads}}>0$):  
  $K_H$ **increases** with $T$.  


**Conclusion:** The Henry coefficient $K_H$ has an **Arrhenius/van ’t Hoff–type dependence** on temperature, governed primarily by $\Delta H_{\mathrm{ads}}$ and $\Delta S_{\mathrm{ads}}$.

# 4 Determining Langmuir Parameters for CO2 Absorption
Step 1: Fitting Isotherms to the Lanmguir Model

The Langmuir absoprtion model is given by:


$$
q(P) = q_{\max}\,\frac{K(T)P}{1+K(T)P}
$$


At 263K, fit the isotherm to the Langmuir equation to detemrine \(q_{sat}\).Indeed, at the lowest temperature (263 K), absorption is strongest which means that the isotherm will reach closest to saturation and we can extract a good estimate of \(q_{sat}\). Then, at all temperatures (263, 303, 343 K), fit the curves again using the same \(q_{sat}\) value but allowing K(T) to vary. This give three temperature-dependant Langmuir constants: K(263), K(303) and K(343). This can be done, since \(q_{sat}\) is independant of temperature, since the MOF structure doesn't change with temperature. 
Graphically: 
 - Plot the experimental absorption isotherms q against P.
 - Fit them with the Langmuir model curves.
 - Extract K(T) for each temperature.

Step 2: Extracting Thermodynamic Parameters with Van't Hoff Equation

From Step 1, we obtained the Langmuir equilibrium constants K(T) at three temperatures.

$$
K(T) = k_0 \, \exp\left(-\frac{\Delta H}{R T}\right)
$$

where:  
- $K(T)$: the equilibrium constant at temperature $T$
- $K_0$: the pre-exponential factor, $K_0 = q_{\max}^0 \exp\left(\frac{\Delta S_\mathrm{ads}}{R}\right)$
- $\Delta H_\mathrm{ads}$: the adsorption enthalpy
- $R$: the universal gas constant
- $T$: the absolute temperature in Kelvin


Taking the natural logarithm gives the van ’t Hoff equation:

$$
\ln K(T) = \ln k_0 - \frac{\Delta H}{R} \cdot \frac{1}{T}
$$



Using the graph:
- On the X-axis: $1/T$ in $K^{-1}$
- On the Y-axis: $\ln K(T)$
- Red points: the three calculated $K(T)$ values
- Blue line: linear regression through the points

From the graph:

1. Slope:
   $$\text{slope} = -\frac{\Delta H}{R}

2. Calculate $\Delta H$:
   
   Using $R = 8.314 \text{ J} \cdot \text{mol}^{-1} \cdot \text{K}^{-1}$
   
   $$\Delta H = -\text{slope} \times R$$

3. Intercept:
   $$\text{intercept} = \ln K_0$$

4. Calculate $K_0$:
   
   Exponentiating gives the pre-exponential factor $K_0$:
   
   $$K_0 = e^{\text{intercept}}$$
