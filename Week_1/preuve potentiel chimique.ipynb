# Proof : Chemical potential and partition function

We want to demonstrate :  

$$
\mu = -k_B T \left(\frac{\partial \ln Q}{\partial N}\right)_{V,T}
$$

---

## 1) Statistical entropy in the canonical ensemble

In the canonical ensemble (where T,V,N are fixed) :

$$
Q(N,V,T) = \sum_{\nu} e^{-\beta E_\nu}, \quad 
\beta = \frac{1}{k_B T}, \quad
p_\nu = \frac{e^{-\beta E_\nu}}{Q}.
$$

Let us begin with the statistical definition of entropy:

$$
S = -k_B \sum_\nu p_\nu \ln p_\nu.
$$

By replacing: $$p_\nu = \frac{e^{-\beta E_\nu}}{Q}$$

We obtain :

$$
\begin{aligned}
S &= -k_B \sum_\nu \frac{e^{-\beta E_\nu}}{Q}\,(-\beta E_\nu - \ln Q) \\
  &= k_B \beta \sum_\nu p_\nu E_\nu + k_B \ln Q \sum_\nu p_\nu \\
  &= \frac{\langle E \rangle}{T} + k_B \ln Q .
\end{aligned}
$$

Since $(\langle E \rangle = E)$ (internal energy in the canonical), we obtain :

$$
TS = E + k_B T \ln Q.
$$

Therefore, Helmholtz's free energy is:

$$
F \equiv E - TS = -k_B T \ln Q.
$$

---

## 2) Thermodynamic identity and chemical potential

Differential identity for $(F(T,V,N))$ is :

$$
dF = -S \, dT - P \, dV + \mu \, dN,
$$

from where :

$$
\mu = \left(\frac{\partial F}{\partial N}\right)_{T,V}.
$$

By using $(F = -k_B T \ln Q)$ (Constant T during differentiation) :

$$
\mu = -k_B T \left(\frac{\partial \ln Q}{\partial N}\right)_{T,V}.
$$

---

## Final result

$$
\boxed{\mu = -k_B T \left(\frac{\partial \ln Q}{\partial N}\right)_{V,T}}
$$
